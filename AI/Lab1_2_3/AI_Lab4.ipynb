5. Agentic functional reparation --> action

[A, Clean] --> Right
[A, Dirty] --> Suck
[B, Clean] --> Left
[B, Dirty] --> Suck

# Simple Reflex

```python
def simple_reflex_agent(percept):
    print('Perception Received: '+ str(percept))
    location = percept[0]
    status = percept[1]
    if status == 'Dirty':
        action = 'Suck'
    elif location == 'A':
        action = 'Right'
    elif location == 'B':
        action = 'Left'
    return action
```

```python
import random
Location = random.choice(['A', 'B'])
Condition = random.choice(['Clean', 'Dirty'])

while True:
    action = simple_reflex_agent((Location, Condition))
    print('Action Performed: ' + action)
    cmd = input('Get Perception (yes/no): ')
    if(cmd == 'no' or cmd!='yes'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean', 'Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['Clean', 'Dirty'])
    else:
        Condition = 'Clean'
```

# Model Based Agents

```python
model = {'A': 'Unknown', 'B': 'Unknown'}
world_state = 'Bad'
action = 'Unknown'

def update_state(action, percept, model):
    location = percept[0]
    status = percept[1]
    model[location] = status # update the model based on current percept
    print('Perception: '+str(model))
    print('Model before action: '+str(model))
    global world_state
    #update the model based on action
    if action == 'Suck':
        model[location] = 'Clean'
    # updatechecking to update world state
    if model['A']=='Clean' and model['B']=='Clean':
        world_state = 'Good'
    else:
        world_state = 'Bad'
    return world_state
```

```python
def model_based_reflex_agent(percept):
    location = percept[0]
    status = percept[1]
    global world_state, action, model
    if world_state == 'Good':
        action = 'Pause'
        return action
    elif status == 'Dirty':
        action = 'Suck'
    elif location == 'A':
        action = 'Right'
    elif location == 'B':
        action = 'left'
    world_state = update_state(action, percept, model)
    print('Action Performed: '+ action)
    print('Model after action: '+ str(model))
    print('State: '+str(world_state))
    return action
```

```python
import random
Location = random.choice(['A', 'B'])
Condition = 'Dirty'

while True:
    action = model_based_reflex_agent((Location, Condition))
    if action == 'Right':
        Location = 'B'
        Condition = 'Dirty'
    elif action == 'Left':
        Location = 'A'
        Condition = 'Dirty'
    elif action == 'Suck':
        Condition = 'Clean'
    elif action == 'Pause':
        cmd = input('Stopped. Do restart? (yes/no): ')
        if(cmd == 'no' or cmd != 'yes'):break
        Location = random.choice(['A', 'B'])
        Condition = random.choice(['Clean', 'Dirty'])
        model = {'A':'Unknown', 'B':'Unknown'}
        world_state = 'Bad'
        action = 'Unknown'
```
