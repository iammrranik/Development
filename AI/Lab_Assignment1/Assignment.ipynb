# Exercise 1: Input a course mark (0-100) from user and use the if statement to print corresponding grade.

```python
userInput = int(input("Input a course mark (0-100): "))
if userInput < 0 or userInput > 100:
    print("Invalid input. Please enter a number between 0 and 100.")
elif userInput >= 90:
    print("Grade: A+")
elif userInput >= 85:
    print("Grade: A")
elif userInput >= 80:
    print("Grade: B+")
elif userInput >= 75:
    print("Grade: B")
elif userInput >= 70:
    print("Grade: C+")
elif userInput >= 65:
    print("Grade: C")
elif userInput >= 60:
    print("Grade: D+")
elif userInput >= 50:
    print("Grade: D")
else:
    print("Grade: F")
```

# Exercise 3:   Print the nth term Fibonacci number
     Sample Input: 10
     Sample Output: 34

```python
def nth_fibonacci(n: int) -> int:
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return nth_fibonacci(n - 1) + nth_fibonacci(n - 2)

userInput = int(input("Input N-th term of Fibonacci sequence (Integer Number): "))
try:
    print(f"The {userInput}th term of the Fibonacci sequence is: {nth_fibonacci(userInput-1)}")
except ValueError:
    print("Please enter a valid integer.")
```

# Exercise 4: Define function which will take two lists of numbers and returns a list (sum of two lists)
    Sample Input: [3,4,5,1] and [6,7,2,8]
    Sample output: [9,11,7,9]

```python
def sum(a: list[int], b: list[int]) -> list[int]:
    temp = []
    if(len(a) == len(b)):
        for i in range(len(a)):
            temp.append(a[i] + b[i])
    elif(len(a) > len(b)):
        for i in range(len(b)):
            temp.append(a[i] + b[i])
        for i in range(len(b), len(a)):
            temp.append(a[i])
    else:
        for i in range(len(a)):
            temp.append(a[i] + b[i])
        for i in range(len(a), len(b)):
            temp.append(b[i])
    return temp

list1 = [1, 2, 3, 4, 5]
list2 = [10, 20, 30]
print(sum(list1, list2))
```

# Exercise 4: Write a function which will take the full name as input and return both the first name and last name.
    Sample Input: Arafat Rahman Sunny
    Sample Output: First name: Arafat, Last name: Sunny

```python
def promt_name(promt: str) -> list[str]:
    while True:
        userInput = input(promt).strip().title()
        if userInput == "":
            print("Empty input is not not allowed. Please, try again.")
            continue
        userInput = userInput.split(" ")
        return userInput

userInput = promt_name("Input your name: ")
if(len(userInput) < 2):
    print("First name: " + userInput[0])
else:
    print("First name: " + userInput[0] + ", Last name: " + userInput[len(userInput) - 1])
```
