# Imports

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
# ১. আপনার আপলোড করা ডেটা লোড করা
df = pd.read_csv('realfishdataset.csv')
```

```python
# ২. অ্যানোমালি লেবেল করা (pH ৬.৫ এর নিচে বা ৮.৫ এর বেশি হলে বিপত্তি)
df['is_anomaly'] = ((df['ph'] < 6.5) | (df['ph'] > 8.5)).astype(int)

X = df[['ph', 'temperature', 'turbidity']]
y = df['is_anomaly']
```

```python
# ৩. মডেল ট্রেনিং
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
# ৪. এক্সটারনাল ডেটা সিমুলেশন (Noise যোগ করা)
X_noisy = X + np.random.normal(0, 0.5, X.shape)
```

```python
# ৫. রেজাল্ট চেক করা (Threshold = 0.7)
probs = model.predict_proba(X_noisy)[:, 1]
preds = (probs >= 0.7).astype(int)

print(f"Accuracy with Noise: {accuracy_score(y, preds)*100:.2f}%")
```
